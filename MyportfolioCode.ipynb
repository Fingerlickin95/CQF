```python
# Import market data module from quantmod
import sys
import os
#sys.path.append(os.path.abspath(".."))

import pandas as pd
from quantmod.markets import getData
```

```python
stocklist = ["SBIN.NS", "ITC.NS", "INFY.NS", "RELIANCE.NS", "TITAN.NS"]
df = getData(stocklist, period="250d")["Close"]
```

```python
# Visualise stock data
import numpy as np
import pandas as pd
import cvxpy as cp

import quantmod.charts
import plotly.graph_objects as go

df.iplot(kind='normalized', showlegend=True)
```

```python
#df of returns and vol

returns = df.pct_change().dropna()
annual_returns = round(returns.mean()*250*100, 2)

annual_stdev = round(returns.std()*np.sqrt(250)*100,2)

#Subsume into df

stats = pd.DataFrame({
    'Ann Ret': annual_returns,
    'Ann Vol': annual_stdev
})

stats.head()
```

```python
# Update quantmod to latest version
# Portfolio composition
stats['Ann Vol'].iplot(kind='pie', showlegend=True, title='Annualised Volatility')
```

```python
#lets compute the stats

mean_returns = (returns.mean()*250).values
cov_matrix = (returns.cov()*250).values

n=len(mean_returns)

Cov_matrix = pd.DataFrame(cov_matrix, index=returns.columns, columns=returns.columns)

Cov_matrix.head()
```

```python
# lets define the formula for our portfolio, this will be a function of wT * Cov_matrix * return matrix
#step 1, we define the weight matrix as a column vector [nx5]

wT = np.array([0.2,0.2,0.3,0.15,0.15])

#[INFY, ITC, RELIANCE, SBIN, TITAN]

if wT.sum() <1:
    difference = 1- wT.sum()
    print("add weight of ", difference )
else:
    print("total portfolio weight is 1")

#step 2 lets build the return matrix

return_matrix = mean_returns


#now we already have the cov matrix from above cell

portfolio_returns = return_matrix @ wT
portfolio_variance = wT @ Cov_matrix @wT.T

print("portfolio return:", portfolio_returns)
print("portfolio_variance:", portfolio_variance)
```
